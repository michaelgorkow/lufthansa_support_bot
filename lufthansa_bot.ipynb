```python
# Imports
from seleniumbase import Driver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
```

```python
# Variables
URL = "https://www.lufthansa.com/de/de/kompensation-bei-flugunregelmaessigkeiten"
BOOKING_CODE = "ABC123"
FLIGHT_NO = "1234"
FLIGHT_DATE = "20.12.2023"
TICKET_NO = "1234567890"
SALUTATION = "Herr"
FIRSTNAME = "firstname"
LASTNAME = "lastname"
MILESMORE_NUMBER = "1234567890"
STREET = "street"
ZIP = "123456"
CITY = "Berlin"
COUNTRY = "Germany"
MAIL = "firstname.lastname@gmail.com"
PHONE = "+49123456789"
CURRENCY = "EUR"
AMOUNT = "123"
sleep_time = 0.25 # adjust if Selenium is too fast
errors = 0 # count errors
NUM_TICKETS = 2
```

```python
for i in range(NUM_TICKETS):
    try:
        browser = Driver(uc_cdp=True, incognito=True)
        browser.get(URL)
        WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="cm-acceptAll"]'))).click()
        tmp = browser.find_elements(By.XPATH, "//span[@class='radio-label-text cmp-form-options__field-description' and text()='Flugausfall']")[0]
        time.sleep(sleep_time)
        tmp.click()
        time.sleep(sleep_time)
        tmp = browser.find_elements(By.XPATH, "//span[@class='radio-label-text cmp-form-options__field-description' and text()='Für mich']")[0]
        time.sleep(sleep_time)
        tmp.click()
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Buchungscode']").send_keys(BOOKING_CODE)
        time.sleep(sleep_time)
        browser.find_element(By.NAME, "carrier_1").send_keys('LU')
        time.sleep(sleep_time)
        input_field = browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Flugnummer']").send_keys(FLIGHT_NO)
        time.sleep(sleep_time)
        input_field = browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Flugdatum']").send_keys(FLIGHT_DATE)
        time.sleep(sleep_time)
        tmp = browser.find_elements(By.XPATH, "//span[@class='radio-label-text cmp-form-options__field-description' and text()='Reise wurde abgebrochen']")[0]
        time.sleep(sleep_time)
        tmp.click()
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Ticketnummer']").send_keys(TICKET_NO)
        time.sleep(sleep_time)
        browser.find_element(By.NAME, "salutation_1").send_keys(SALUTATION)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Vorname']").send_keys(FIRSTNAME)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Name']").send_keys(LASTNAME)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Miles & More Nummer']").send_keys(MILESMORE_NUMBER)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Straße, Hausnummer']").send_keys(STREET)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Postleitzahl']").send_keys(ZIP)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Stadt']").send_keys(CITY)
        time.sleep(sleep_time)
        browser.find_element(By.NAME, "country_1").send_keys(COUNTRY)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='E-Mail-Adresse']").send_keys(MAIL)
        time.sleep(sleep_time)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Telefonnummer']").send_keys(PHONE)
        browser.find_element(By.NAME, "currency").send_keys(CURRENCY)
        browser.find_element(By.XPATH, "//input[@data-cmp-hook-form-text='input' and @placeholder='Gesamtkosten']").send_keys(AMOUNT)
        time.sleep(sleep_time)
        browser.find_elements(By.XPATH, "//span[@class='custom-control-description' and text()='Ich bestätige, dass die Angaben im obigen Formular der Wahrheit entsprechen.']")[0].click()
        time.sleep(sleep_time)
        browser.find_elements(By.XPATH, "//span[@class='custom-control-description' and text()='Ich akzeptiere die Datenschutzbestimmungen.']")[0].click()
        time.sleep(sleep_time)
        #browser.find_element(By.XPATH,"//button[@type='submit']").click()
        print(f'Submitted ticket{i}')
        time.sleep(1)
        browser.quit()
    except Exception as e:
        print(e)
        errors += 1
        print(f'{i} did not work. Error:{errors}')
        browser.quit()
```
